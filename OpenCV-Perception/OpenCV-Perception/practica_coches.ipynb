```python
# leer y vizualizar el video
import cv2

cap = cv2.VideoCapture("imgs/carretera.mp4")
frames = []

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    frames.append(frame)
    cv2.imshow("Video", frame)
    cv2.waitKey(25)

cap.release()
cv2.destroyAllWindows()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

frames_array = np.array(frames, dtype=np.float32)
img_fondo = np.mean(frames_array, axis=0)
img_fondo = img_fondo.astype(np.uint8)

plt.imshow(cv2.cvtColor(img_fondo, cv2.COLOR_BGR2RGB))  
plt.title("Imagen fondo")
plt.axis('off')
plt.show()
```

```python
T = 50 # Umbral

for i, frame in enumerate(frames):
    sub = cv2.absdiff(frame, img_fondo)
    
    sub_gray = cv2.cvtColor(sub, cv2.COLOR_BGR2GRAY)
    
    _, image_bin = cv2.threshold(sub_gray, T, 255, cv2.THRESH_BINARY)
    
    plt.imshow(image_bin, cmap='gray')
    plt.title(f"Frame {i+1} - Imagen binarizada (T={T})")
    plt.axis('off')
    plt.tight_layout()
    #Splt.show()
    plt.show()
```
